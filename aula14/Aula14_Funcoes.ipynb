*Aquecimento*: Fa√ßa uma fun√ß√£o que recebe tr√™s par√¢metros. O primeiro √© um caractere que indica a opera√ß√£o: +, -, / e *. Os outros dois s√£o valores inteiros. A fun√ß√£o deve retornar o valor da opera√ß√£o sobre os dois valores, ou seja: n1 ‚Äúop‚Äù n2 ÔÉ† Fun√ß√£o: calcula(‚Äúop‚Äù, n1, n2)

```python
def calcula(op = "+", n1, n2):
    if op == "+":
        r = n1 + n2
    elif op == "-":
        r = n1 - n2
    elif op == "*":
        r = n1 * n2
    elif op == "/":
        r = n1 / n2
    else:
        r = -1
    return int(r)
```

```python
# Teste da Fun√ß√£o Calcula()
a = calcula("+", 7, 8)
b = calcula("-", 8, 2)
c = calcula("*", 2, 3)
d = calcula("/", 10, 5)
print(a, b, c, d)
# resultado esperado: 15 6 6 2
```

```python
def central_respostas(*args):
    if args:
        qtd = len(args)
        print(qtd, "mensagem(ns) recebida(s).")
        i = 1
        for msg in args:
            print(f"Mensagem {i} recebida: {msg}")
            i = i + 1
    else:
        print("Nenhuma mensagem recebida. Por favor, tente novamente.")
```

```python
central_respostas(
    "Gostaria de saber o status do meu pedido.",
    "Tamb√©m quero alterar meu endere√ßo de entrega."
)
# Sa√≠da:
# 2 mensagens recebidas.
# Mensagem 1 recebida: Gostaria de saber o status do meu pedido.
# Mensagem 2 recebida: Tamb√©m quero alterar meu endere√ßo de entrega.
```

```python
# Fun√ß√£o para somar n√∫meros
def addNum(n1, n2):
    print("Primeiro n√∫mero: " + str(n1))
    print("Segundo n√∫mero: " + str(n2))
    print("Soma: ", n1 + n2)
```

```python
# Chamando a fun√ß√£o e passando par√¢metros
addNum(4, 7)
```

```python
# Chamando a fun√ß√£o e passando par√¢metros
addNum(45, 3)
```

```python
# Fun√ß√µes com n√∫mero vari√°vel de argumentos
# √â muito comum encontrar como nome dessa vari√°vel: *args
def printVarInfo( arg1, *vartuple ):
    
   # Imprimindo o valor do primeiro argumento
    print ("O par√¢metro passado foi: ", arg1)
   
   # Imprimindo o valor do segundo argumento 
    for item in vartuple:
        print ("O par√¢metro passado foi: ", item)
    return;
```

```python
# Fazendo chamada √† fun√ß√£o usando apenas 1 argumento
printVarInfo(10)
```

```python
printVarInfo('Chocolate', 'Morango')
```

```python
printVarInfo('PyPRO', 'Fatec', 'CDN', 'Ci√™ncia de Dados')
```

```python
import math

# Verificando se um n√∫mero √© primo
def numPrimo(num):
    if (num % 2) == 0 and num > 2: 
        return "Este n√∫mero n√£o √© primo"
    for i in range(3, int(math.sqrt(num)) + 1, 2):
        if (num % i) == 0:
            return "Este n√∫mero n√£o √© primo"
    return "Este n√∫mero √© primo"
```

1. Fa√ßa um programa que use a fun√ß√£o valorPagamento para determinar o valor a ser pago por uma presta√ß√£o de uma conta. O programa dever√° solicitar ao usu√°rio o valor da presta√ß√£o e o n√∫mero de dias em atraso e passar estes valores para a fun√ß√£o valorPagamento, que calcular√° o valor a ser pago e devolver√° este valor ao programa que a chamou. O programa dever√° ent√£o exibir o valor a ser pago na tela. Ap√≥s a execu√ß√£o o programa dever√° voltar a pedir outro valor de presta√ß√£o e assim continuar at√© que seja informado um valor igual a zero para a presta√ß√£o. Neste momento o programa dever√° ser encerrado, exibindo o relat√≥rio do dia, que conter√° a quantidade e o valor total de presta√ß√µes pagas no dia. O c√°lculo do valor a ser pago √© feito da seguinte forma. Para pagamentos sem atraso, cobrar o valor da presta√ß√£o. Quando houver atraso, cobrar 3% de multa, mais 0,1% de juros por dia de atraso

```python
def valorPagamento(valor, diasAtraso):
    if (valor < 0):
        return None
    if (diasAtraso > 0):
        multa = valor * 0.03
        adicionalAtraso = valor * (diasAtraso * 0.01)
        return valor + multa + adicionalAtraso
    else:
        return valor
```

```python
# Entrada de Dados
valor = 1
while (valor != 0):
    valor = float(input('Informe o valor da presta√ß√£o: '))
    if (valor != 0):
        diasAtraso = int(input('Informe a qtd de dias de atraso: '))
        print(f"Valor a ser pago: {valorPagamento(valor, diasAtraso):.2f}")
```

2. Embaralha palavra. Construa uma fun√ß√£o que receba uma string como par√¢metro e devolva outra string com os carateres embaralhados. Por exemplo: se fun√ß√£o receber a palavra python, pode retornar npthyo, ophtyn ou qualquer outra combina√ß√£o poss√≠vel, de forma aleat√≥ria. Padronize em sua fun√ß√£o que todos os caracteres ser√£o devolvidos em caixa alta ou caixa baixa, independentemente de como foram digitados

```python
import random

def embaralhaPalavra(palavra):
    arrayPos = []
    tamanho = len(palavra)
    for i in range(0, tamanho):
        arrayPos.append(i)
    for i in range(0, tamanho):
        pos1 = random.randint(0, tamanho - 1)
        pos2 = random.randint(0, tamanho - 1)

        aux = arrayPos[pos1]
        arrayPos[pos1] = arrayPos[pos2]
        arrayPos[pos2] = aux

    retorno = ''
    for i in arrayPos:
        retorno += palavra[i]

    return retorno.upper()
```

```python
# Fluxo Principal
palavra = input('Informe uma palavra: ')
print(embaralhaPalavra(palavra))
```

Fa√ßa um jogo completo ‚Äúpedra ‚Äì papel ‚Äì tesoura‚Äù.  
A pessoa joga contra o computador. Ao iniciar, pergunta-se o nome do jogador.  
Ao final de cada Jogo, pergunta se quer continuar jogando.  
Ao finalizar, depois de ‚Äún‚Äù jogadas, apresenta-se o resultado:  
Vit√≥rias de nome....: 3  
Vit√≥rias do Computador: 2  
Empates...............: 4   
nome foi o grande Campe√£o! 

```python
from random import choice

# 1. Inicializa√ß√£o
nome_jogador = input("Qual √© o seu nome? ")
opcoes = ["pedra", "papel", "tesoura"]

vitorias_jogador = 0
vitorias_computador = 0
empates = 0

print(f"\nSeja bem-vindo(a), {nome_jogador}! Vamos jogar Pedra, Papel, Tesoura!")

# 2. Loop Principal
while True:
    print("\n--- Nova Rodada ---")
    
    # 2.1 Escolha do Jogador (Com Valida√ß√£o)
    while True:
        escolha_jogador = input("Escolha (pedra, papel, tesoura) ou 'sair' para terminar: ").lower()
        if escolha_jogador in opcoes:
            break
        elif escolha_jogador == 'sair':
            break
        else:
            print("Escolha inv√°lida. Tente novamente.")
            
    if escolha_jogador == 'sair':
        break
        
    # 2.2 Escolha do Computador
    escolha_computador = choice(opcoes)
    
    # 2.3 Apresentar as Escolhas
    print(f"\n{nome_jogador} escolheu: **{escolha_jogador.upper()}**")
    print(f"O Computador escolheu: **{escolha_computador.upper()}**")
    
    # 2.4 Determinar o Vencedor
    if escolha_jogador == escolha_computador:
        resultado = "Empate!"
        empates += 1
    elif (escolha_jogador == "pedra" and escolha_computador == "tesoura") or \
         (escolha_jogador == "papel" and escolha_computador == "pedra") or \
         (escolha_jogador == "tesoura" and escolha_computador == "papel"):
        resultado = f"Vit√≥ria para {nome_jogador}!"
        vitorias_jogador += 1
    else:
        resultado = "Vit√≥ria para o Computador!"
        vitorias_computador += 1
        
    print(f"**{resultado}**")
    
    # 2.5 Apresentar o Placar Parcial
    print(f"\n--- PLACAR ATUAL ---")
    print(f"| {nome_jogador}: {vitorias_jogador} | Computador: {vitorias_computador} | Empates: {empates} |")
    
    # 2.6 Perguntar se Quer Continuar
    continuar = input("\nDeseja jogar novamente? (s/n): ").lower()
    if continuar != 's':
        break

# 3. Resultado Final
print("\n" + "="*30)
print("     FIM DE JOGO!")
print("="*30)

print(f"Vit√≥rias de {nome_jogador}: {vitorias_jogador}")
print(f"Vit√≥rias do Computador: {vitorias_computador}")
print(f"Empates...............: {empates}")
print("-"*30)

# Declarar o Campe√£o
if vitorias_jogador > vitorias_computador:
    print(f"**{nome_jogador} foi o grande Campe√£o!** üèÜ")
elif vitorias_computador > vitorias_jogador:
    print("**O Computador foi o grande Campe√£o desta vez!** üíª")
else:
    print("**Houve um empate no placar final! Jogo acirrado!**")
    
```
